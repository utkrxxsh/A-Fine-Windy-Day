# * A Fine Windy Day

```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('X_train.csv')
```

```python
train.isnull().sum().sum()
```

```python
train=train.fillna(0)
```

```python
X_train=train.iloc[:,1:]
y_train=train.iloc[:,0]
X_test=pd.read_csv('X_test.csv')
```

```python
X_test.isnull().sum().sum()
```

```python
X_test=X_test.fillna(0)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
ridge_reg=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_reg.fit(X_train,y_train)
```

```python
y_pred_ridge=ridge_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
y_pred=ridge_reg.predict(X_train)
print(r2_score(y_train,y_pred))
```

```python
y_pred_ridge
```

```python
org=pd.read_csv('test.csv')
rem=org.iloc[:,:2]
y_pred_ridge=pd.DataFrame(y_pred_ridge)
y_pred_ridge.rename(columns={0:'windmill_generated_power(kW/h)'},inplace=True)
final_ans=pd.concat([rem, y_pred_ridge], axis=1)
```
