# * A Fine Windy Day

```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('X_train.csv')
train=train.fillna(0)
X_train=train.iloc[:,1:]
y_train=train.iloc[:,0]
X_test=pd.read_csv('X_test.csv')
X_test=X_test.fillna(0)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
lasso_reg=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_reg.fit(X_train,y_train)
```

```python
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```

```python
prediction = lasso_reg.predict(X_test)
```

```python
y_pred=lasso_reg.predict(X_train)
from sklearn.metrics import r2_score
print(r2_score(y_train,y_pred))
```

```python
prediction
```

```python
org=pd.read_csv('test.csv')
rem=org.iloc[:,:2]
prediction=pd.DataFrame(prediction)
prediction.rename(columns={0:'windmill_generated_power(kW/h)'},inplace=True)
final_ans=pd.concat([rem, prediction], axis=1)
```
